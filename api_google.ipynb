```python
import os
import io
from google.cloud import vision

# üîπ Configurar credenciales (escapa las barras invertidas o usa una cadena sin formato)
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"C:\Users\antony.portocarrer\OneDrive - Interpublic\Documents\ee-python\flash-district-452304-i1-91230a8863dd.json"

# üîπ Crear cliente de Vision API
client = vision.ImageAnnotatorClient()

# üîπ Ruta de la imagen (aj√∫stala seg√∫n tu archivo real)
image_path = r"C:\Users\antony.portocarrer\OneDrive - Interpublic\Documents\ee-python\image_3.jpg"

# üîπ Cargar la imagen y enviarla a la API
try:
    with io.open(image_path, 'rb') as image_file:
        content = image_file.read()

    image = vision.Image(content=content)
    response = client.text_detection(image=image)

    # üîπ Verificar si hay texto detectado
    texts = response.text_annotations
    if texts:
        extracted_text = texts[0].description
        print("Texto extra√≠do:\n", extracted_text)
        
        # üîπ Guardar en un archivo TXT
        output_path = r"C:\Users\antony.portocarrer\OneDrive - Interpublic\Documents\ee-python\texto_extraido.txt"
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(extracted_text)
        print(f"‚úÖ Texto guardado en '{output_path}'")
    else:
        print("‚ö†Ô∏è No se detect√≥ texto en la imagen.")

except Exception as e:
    print(f"‚ùå Error: {e}")
```

```python
import os
import io
import pandas as pd
from google.cloud import vision

# Configurar credenciales
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"C:\Users\antony.portocarrer\OneDrive - Interpublic\Documents\ee-python\flash-district-452304-i1-8cfad4d5c8ae.json"

# Crear cliente de Vision API
client = vision.ImageAnnotatorClient()

# Carpeta con im√°genes
input_folder = r"C:\Users\antony.portocarrer\OneDrive - Interpublic\Documents\ee-python\imagenes_productos"
output_file = r"C:\Users\antony.portocarrer\OneDrive - Interpublic\Documents\ee-python\textos_extraidos.xlsx"

# Lista para almacenar resultados
data = []

# Procesar cada imagen en la carpeta
for filename in os.listdir(input_folder):
    if filename.lower().endswith((".jpg", ".png", ".jpeg")):  # Soporta formatos de imagen
        image_path = os.path.join(input_folder, filename)

        with io.open(image_path, 'rb') as image_file:
            content = image_file.read()

        image = vision.Image(content=content)
        response = client.text_detection(image=image)
        texts = response.text_annotations

        extracted_text = texts[0].description if texts else "No se detect√≥ texto"
        data.append([filename, extracted_text])  # Agrega a la lista

        print(f"‚úÖ Procesado: {filename}")

# Convertir a DataFrame y guardar en Excel
df = pd.DataFrame(data, columns=["Nombre de la Imagen", "Texto Extra√≠do"])
df.to_excel(output_file, index=False)

print(f"\nüìÇ Archivo Excel guardado en: {output_file}")
print(f"üìä Total im√°genes procesadas: {len(data)}")
```
